```python
tesla_2024 = yf.download("TSLA", start = '2017-05-28', end = dt.datetime.now())
```

```python
tesla_2024 = tesla_2024.reset_index()
tesla_2024 = tesla_2024[['Date','Open','High','Low','Close']]
print(tesla_2024.shape)
tesla_2024.head()
```

```python
tesla_2024['month'] = pd.DatetimeIndex(tesla_2024['Date']).month
tesla_2024['year'] = pd.DatetimeIndex(tesla_2024['Date']).year
tesla_2024.head()
```

```python
teslaPivot = pd.pivot_table(tesla_2024, values="Close", columns="year", index="month")
teslaPivot.head()
```

```python
import pandas as pd
import numpy as np

# Assuming 'data' contains the table you provided
data = {
    'year': [2011, 2011, 2011, 2012, 2012, 2012],
    'month': [1, 2, 3, 1, 2, 3],
    'value': [27.49, 35.18, 164.05, 203.25, 204.56, 239.32]
}

df = pd.DataFrame(data)

# Create a pivot table
pivot_table = pd.pivot_table(df, values='value', index='year', columns='month', aggfunc=np.sum)

print(pivot_table)
```
