```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from yahooquery import Ticker
import yfinance as yf
```

```python
df = yf.download('^AEX', multi_level_index= False)

df
```

```python
import mplfinance as mpf

mpf.plot(df,type='candle', figsize=(15,5))
```

```python
import finplot as fplt
```

```python
# PLOT THE OHLC CANDLE CHART
fplt.candlestick_ochl(df[['Open','Close','High','Low']])
fplt.show()
```

```python
mpf.plot(df['2020-01-01':], type='ohlc', mav=4, figsize=(13,5))
```

```python
mpf.plot(df['2020-01-01':], type='candle', mav=(5, 25, 75), style='yahoo', axtitle='Itel')
```

```python
mpf.plot(df['2020-01-01':], type='pnf', figsize=(13,5))
```

```python
mpf.plot(df['2023-01-01':], type='renko', figsize=(13,5))
```
